```python
import pandas as pd

# Cargar los archivos
archivo_principal = 'ensadul2021_df_decode.csv'
archivo_secundario = 'ensadul2021_int_df_decode.csv'

# Leyendo de nuevo los archivos usando ';' como delimitador
df_principal = pd.read_csv(archivo_principal, delimiter=';',low_memory=False)
df_secundario = pd.read_csv(archivo_secundario, delimiter=';',low_memory=False)


# Revisando las primeras filas para confirmar la lectura correcta
df_principal.head(), df_secundario.head()
```

```python
# Renombrando las columnas del segundo dataset añadiendo el prefijo "INT_"
df_secundario_renombrado = df_secundario.add_prefix('INT_')

# Uniendo los archivos con la indicación de si hay coincidencia
df_unido_con_indicador = pd.merge(
    df_principal, 
    df_secundario_renombrado, 
    left_on=['Folio', 'Folio de integrante'], 
    right_on=['INT_Folio', 'INT_Folio de integrante'], 
    how='left', 
    indicator=True
)

# Creando una columna que indica si hay coincidencia (True/False)
df_unido_con_indicador['Coincidencia'] = df_unido_con_indicador['_merge'] == 'both'

# Eliminando la columna '_merge' y las columnas duplicadas de 'Folio' y 'Folio de integrante' del segundo dataset
df_unido_con_indicador = df_unido_con_indicador.drop(columns=['_merge'])

df_unido_con_indicador.head()
```

```python
print(df_principal.shape[0])

print(df_secundario.shape[0])

print(df_unido_con_indicador.shape[0])
```

```python

# Guardando el dataset con nuevos encabezados en un archivo CSV
df_unido_con_indicador.to_csv('ensadul2021_Int_union.csv',sep=';', index=False)
```

```python
# Cargando el tercer archivo
archivo_tercero = 'ensadul2021_antro_df_decode.csv'

df_tercero = pd.read_csv(archivo_tercero, delimiter=';',low_memory=False)


# Revisando las primeras filas para confirmar la lectura correcta
df_tercero.head()

```

```python

# Renombrando las columnas del tercer dataset añadiendo el prefijo "ANTRO_"
df_tercero_renombrado = df_tercero.add_prefix('ANTRO_')

# Revisando las primeras filas para confirmar la lectura correcta
df_tercero_renombrado.head()
```

```python

# Uniendo el dataset df_unido_con_indicador con el tercer dataset renombrado
df_unido_total = pd.merge(
    df_unido_con_indicador, 
    df_tercero_renombrado, 
    left_on=['Folio', 'Folio de integrante'], 
    right_on=['ANTRO_Folio', 'ANTRO_Folio de integrante'], 
    how='left', 
    indicator='coincidencia_Antro'
)

# Creando una columna que indica si hay coincidencia en el tercer archivo (True/False)
df_unido_total['coincidencia_Antro'] = df_unido_total['coincidencia_Antro'] == 'both'


df_unido_total.head()
```

```python
print(df_principal.shape[0])

print(df_secundario.shape[0])

print(df_unido_con_indicador.shape[0])

print(df_tercero.shape[0])

print(df_unido_total.shape[0])
```

```python

# Guardando el dataset con nuevos encabezados en un archivo CSV
df_unido_total.to_csv('ensadul2021_Int_antro_union.csv',sep=';', index=False)
```

```python
# Revisando si hay registros en el dataset unido que no tienen coincidencia
registros_sin_coincidencia = df_unido_total[df_unido_total['coincidencia_Antro'] == False]

# Contando cuántos registros no tienen coincidencia
num_registros_sin_coincidencia = len(registros_sin_coincidencia)
num_registros_sin_coincidencia

print(registros_sin_coincidencia)
```
