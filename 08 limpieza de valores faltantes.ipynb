```python
import pandas as pd

# Carga del dataset con las columnas seleccionadas
archivo = 'ensadul2021SelcolOH.csv'
data = pd.read_csv(archivo, delimiter=';')

# mostrando los primeros renglones del dataset
data.head()
```

```python
resumen = data.describe()

resumen
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribución de los atributos numéricos
numerical_data2 = data.select_dtypes(include=['float64', 'int64'])

num_columns2 = len(numerical_data2.columns)

# Graficando histogramas de datos numéricos
plt.figure(figsize=(15, num_columns2 * 3))
for i, col in enumerate(numerical_data2.columns):
    plt.subplot(num_columns2, 1, i + 1)
    sns.histplot(numerical_data2[col], kde=False, bins=20)
    plt.title(col)
    plt.tight_layout()

# Desplegando los histogramas
plt.show()
```

```python
# Analizando datos faltantes
datos_faltantes = data.isnull().sum()
porcentaje_datos_faltantes = (datos_faltantes / len(data)) * 100

analisis_datos_faltantes = pd.DataFrame({'Datos Faltantes': datos_faltantes, 'Porcentaje': porcentaje_datos_faltantes})

# Mostrando solo columnas con datos faltantes
analisis_datos_faltantes = analisis_datos_faltantes[analisis_datos_faltantes['Datos Faltantes'] > 0]
analisis_datos_faltantes.sort_values(by='Porcentaje', ascending=False)
```

```python
# Mostrando la estructura antes de borrar columnas con valores faltantes
data.shape
```

```python
# descartando las columnas que tienen mas del 70% de los datos perdidos 
limite = 70  # percentage
columnas_descartar = analisis_datos_faltantes[analisis_datos_faltantes['Porcentaje'] > limite].index

# borrando columnas descartadas del dataset
data_l1 = data.drop(columns=columnas_descartar)

# Mostrando la estructura despues del borrado de columnas con datos faltantes
data_l1.shape

```

```python
# Identificando columnas con dos o tres valores únicos
columnas_valores_unicos = data_l1.columns[data_l1.nunique() <= 3].tolist()

columnas_valores_unicos

```

```python

# Aplicando imputación por la moda a estas columnas
for column in columnas_valores_unicos:
    mode_value = data_l1[column].mode()[0]
    data_l1[column].fillna(mode_value, inplace=True)

# Verificando si aún hay valores faltantes en estas columnas
data_check1 = data_l1[columnas_valores_unicos].isnull().sum()
data_check1.any()  # Retorna False si no quedan valores faltantes



```

```python
# Idenficando columnas numéricas en el dataset
columnas_numericas = data_l1.select_dtypes(include=['int64', 'float64']).columns

# aplicando la media para rellenar los valores faltates
data_l2 = data_l1.copy()
for column in columnas_numericas:
    mean_value =  round(data_l1[column].mean())
    data_l2[column].fillna(mean_value, inplace=True)

# validando que ya no haya valores faltantes
valores_faltantes_check = data_l2[columnas_numericas].isnull().sum()
valores_faltantes_check.any()  # Returns False if no missing values are left
```

```python
# Guardando el dataset limpio de valores faltantes
data_l2.to_csv('ensadul2021_l2.csv',sep=';', index=False)
```
