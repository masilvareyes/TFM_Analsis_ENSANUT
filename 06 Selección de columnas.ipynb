```python
import pandas as pd

# Cargando el listado de columnas seleccionado
file_path = 'ColSeleccion.xlsx'
selectcol = pd.read_excel(file_path)

# validando carga
selectcol.head()
```

```python
# Cargar el archivo en un DataFrame de pandas
file_path = 'ensadul2021_Limpio.csv'
ensadul2021 = pd.read_csv(file_path, sep=';',low_memory=False)

# Mostrar las primeras filas del DataFrame para verificar su contenido
ensadul2021.head()
```

```python
# seleccionando las columnas del archivo 'ColSeleccion.xlsx' para filtrar el archivo origen 'ensadul2021_Limpio.csv'
selectcolls = selectcol['Descripcion'].tolist()
selectcolls
```

```python

# Extrayendo las columnas seleccionadas del dataset'ColSeleccion.xlsx'
filtered_csv_data = ensadul2021[selectcolls]
```

```python
# mostrando para validar
filtered_csv_data.head()
```

```python
# Guardando el dataset con nuevos encabezados en un archivo CSV
filtered_csv_data.to_csv('ensadul2021Selcol.csv',sep=';', index=False)
```
