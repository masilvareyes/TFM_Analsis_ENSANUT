```python
import pandas as pd

# archivo
archivo = 'temp7.csv'

# carga del archivo
data = pd.read_csv(archivo, sep=';')

data_seleccion = data.iloc[:, 5:]


# creando un dataframe con los valores unicos de cada columna/atributo
data_valores_unicos_df = pd.DataFrame({col: pd.Series(data_seleccion[col].unique()) for col in data_seleccion.columns})


data_valores_unicos_df
```

```python
# Guardar en archivo
archivo_unicos = 'data_valores_unicos7.csv'
data_valores_unicos_df.to_csv(archivo_unicos, sep=';', index=False)
```
