```python
import pandas as pd

ensadul2021_int= "data/integrantes_ensanut2021_w_12_01_2022.csv"
catalogo_int= "data/integrantes_ensanut2021_w_12_01_2022.Catálogo.xlsx"
claves_int= 'data/claves_integrantes.xlsx'
```

```python
# Cargar el archivo de datos ensadul2021
ensadul2021_int_df = pd.read_csv(ensadul2021_int, sep=';',low_memory=False)

ensadul2021_int_df
```

```python
# Cargar el archivo de claves
claves_int_df = pd.read_excel(claves_int)

# Verificando las primeras filas para entender la estructura del dataset
claves_int_df.head()

```

```python
# b Creando datasets separados para cada valor único en la columna 'Atributo' de claves
unique_attributes = claves_int_df['Atributo'].unique()
separated_datasets = {attr: claves_int_df[claves_int_df['Atributo'] == attr] for attr in unique_attributes}

# Verificando los nombres de los datasets creados
list(separated_datasets.keys())
```

```python
# c Preparando para realizar las sustituciones en el dataset 'ensadul2021'

# Crear un diccionario para cada atributo con sus valores y etiquetas correspondientes
replacement_dict = {attr: separated_datasets[attr].set_index('Valor')['Etiqueta'].to_dict() 
                    for attr in separated_datasets}

replacement_dict





```

```python
replacement_dict_ent = {}
for k, v in replacement_dict.items():
    replacement_dict_ent[k] = {}
    for sub_k, sub_v in v.items():
        try:
            float_key = float(sub_k)
            int_key = int(float_key)
            replacement_dict_ent[k][int_key] = sub_v
        except ValueError:
            # Maneja claves que no son numéricas
            replacement_dict_ent[k][sub_k] = sub_v

# Imprimir el resultado
replacement_dict_ent
```

```python
replacement_dict_txt = {k: {str(sub_k): sub_v for sub_k, sub_v in v.items()} for k, v in replacement_dict_ent.items()}

replacement_dict_txt
```

```python

# Iterar sobre cada columna en 'ensadul2021_df' y realizar las sustituciones
for col in ensadul2021_int_df.columns:
    if col in replacement_dict_txt:
        # Reemplazando los valores en la columna con las etiquetas correspondientes del diccionario 'replacement_dict_txt'
        ensadul2021_int_df[col] = ensadul2021_int_df[col].astype(str).map(replacement_dict_txt[col]).fillna(ensadul2021_int_df[col])
        print(ensadul2021_int_df[col].head())

# Verificando las primeras filas después de las sustituciones
print(ensadul2021_int_df.head())
```

```python
###########################################################################
####################====== SOLO PARA DEBUG ======##########################
###########################################################################



# Imprimir el diccionario de reemplazo para la columna 'x_region'
print("Diccionario de reemplazo para 'x_region':", replacement_dict_txt.get('x_region', 'No encontrado'))

# Comprobar el tipo de datos de los valores en la columna 'x_region'
print("Tipo de datos en 'x_region':", ensadul2021_int_df['x_region'].dtype)

# Realizar el reemplazo para la columna 'x_region' y verificar los resultados
if 'x_region' in replacement_dict_txt:
    # Utiliza el mapeo con el diccionario de reemplazo para 'x_region'
    ensadul2021_int_df['x_region'] = ensadul2021_int_df['x_region'].astype(str).map(replacement_dict_txt['x_region']).fillna(ensadul2021_int_df['x_region'])
    print("Primeras filas después del reemplazo en 'x_region':")
    print(ensadul2021_int_df['x_region'].head())
  
    
else:
    print("La columna 'x_region' no está en el diccionario de reemplazo.")
```

```python
# e Guardando el dataset pivotado en un archivo CSV
ensadul2021_int_df.to_csv('ensadul2021_int_decode.csv',sep=';', index=False)
```
